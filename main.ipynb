```python
from liquidity_pool import LiquidityPool
from insure_pool import InsurePool
from wallet import Wallet

lp = LiquidityPool(
    asset_reserve=1000,
    base_reserve=1000
)

ip = InsurePool(
    lp=lp,
)

wallet_a = Wallet(address="0xA", base=200)
wallet_b = Wallet(address="0xB", base=200)
wallet_c = Wallet(address="0xC", base=200)

buyer_one_amount = 180
buyer_two_amount = 100

ip.buy(wallet_a, buyer_one_amount, 0)
ip.buy(wallet_b, buyer_two_amount, 0)
ip.buy(wallet_c, buyer_two_amount, 0)

print("uninsured example")
wallet_a.print_balances()
wallet_b.print_balances()
wallet_c.print_balances()

ip.sell(wallet_a)
ip.sell(wallet_b)
ip.sell(wallet_c)

print()

wallet_a.print_balances()
wallet_b.print_balances()
wallet_c.print_balances()

print()

# Wallet 0xA has 20 base and 152.54 asset.
# Wallet 0xB has 100 base and 66.21 asset.
# Wallet 0xC has 100 base and 56.61 asset.

# Wallet 0xA has 259.98 base and 0.0 asset.
# Wallet 0xB has 180.01 base and 0.0 asset.
# Wallet 0xC has 160.01 base and 0.0 asset.

lp = LiquidityPool(
    asset_reserve=1000,
    base_reserve=1000
)

ip = InsurePool(
    lp=lp,
)

wallet_a = Wallet(address="0xA", base=200)
wallet_b = Wallet(address="0xB", base=200)
wallet_c = Wallet(address="0xC", base=200)

insurance = 10

ip.buy(wallet_a, buyer_one_amount, insurance)
ip.buy(wallet_b, buyer_two_amount, insurance)
ip.buy(wallet_c, buyer_two_amount, insurance)

print("insured example")
wallet_a.print_balances()
wallet_b.print_balances()
wallet_c.print_balances()

ip.sell(wallet_a)
ip.sell(wallet_b)
ip.sell(wallet_c)

print()

wallet_a.print_balances()
wallet_b.print_balances()
wallet_c.print_balances()

# Wallet 0xA has 10 base and 152.54 asset.
# Wallet 0xB has 90 base and 66.21 asset.
# Wallet 0xC has 90 base and 56.61 asset.

# Wallet 0xA has 249.98 base and 0.0 asset.
# Wallet 0xB has 185.01 base and 0.0 asset.
# Wallet 0xC has 165.01 base and 0.0 asset.
```
